```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# Loading dataset to a pandas dataframe
df=pd.read_csv('creditcard.csv')
```

```python
# 5 random rowss of dataset
df.tail()
```

```python
# dataset information
df.info()
```

```python
# checking the no. of missing values in each column
df.isnull().sum()
```

```python
# distribution of legit transcation and  fraudlent transc.
df['Class'].value_counts()
```

```python
# So here we can see that legit transc are more than fraudlent one so whenever we give any test data it will predict it as. normal transc .
# This is unbalanced dataset and we cannot train our machine learning model in it
# 0    284315 (Normal Transc.)
# 1       492 (Fraud Transc.)
```

```python
# Seprating the data for analysis
legit=df[df.Class==0]
fraud=df[df.Class==1]
```

```python
fraud.shape
```

```python
# Satatical mearsure of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# Compare the values for both transc
df.groupby('Class').mean()
```

```python
# Under Sampling
# Build sample dataset containg similar distribution of legit transc. and fraud transc.
```

```python
legitSample=legit.sample(n=492)
```

```python
legitSample
```

```python
# Concatenating two dataframe
```

```python
newDF=pd.concat([legitSample,fraud],axis=0) # axis=0 (row wise)   and axis=1 (column wise)
```

```python
newDF
```

```python
newDF['Class'].value_counts()
```

```python
newDF.groupby("Class").mean()
```

```python
# Splitting the data into features and targets
```

```python
X=newDF.drop(columns='Class',axis=1)
y=newDF['Class']
```

```python
X
```

```python
y
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.2,stratify=y,random_state= 2)
```

```python
X_train.shape 
```

```python
X_test.shape
```

```python
# Model Training
```

```python
model=LogisticRegression()
```

```python
# Training the data in lr model with training data
model.fit(X_train,y_train)
```

```python
# Evaluation 
# Accuracy score
```

```python
# Accuracy on training data
xTrainPred=model.predict(X_train)
trainingAcc=accuracy_score(xTrainPred,y_train)
```

```python
trainingAcc
```

```python
xTestPred=model.predict(X_test)
testingAcc=accuracy_score(xTestPred,y_test)
```
